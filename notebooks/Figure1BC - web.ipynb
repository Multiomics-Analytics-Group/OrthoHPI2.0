```python
import os
import yaml
import pandas as pd
import plotly.express as px
```

```python
data_dir_path = '../data'
data_path = os.path.join(data_dir_path, 'predictions.tsv')
config_path = os.path.join('..', 'config.yml')
```

```python
with open(config_path, 'r') as stream:
    content = yaml.safe_load(stream)
```

```python
predictions = pd.read_csv(data_path, sep='\t', header=0)
```

```python
predictions.head()
```

```python
predictions.shape
```

```python
predictions.drop_duplicates(['source', 'target'])
```

```python
predictions.taxid1.unique().shape
```

```python
predictions.taxid2.unique().shape
```

```python
predictions[predictions['edge_type'] == 'inter-species'].shape
```

```python
predictions[(predictions['average_score'] >0.7) & (predictions['taxid1'] != predictions['taxid2'])].shape
```

```python
predictions[predictions['taxid1'].isin([5660, 6185, 6182, 6248])]
```

```python
net = predictions[predictions['taxid1'] != predictions['taxid2']]
```

```python
net.shape
```

```python
net.head()
```

```python
labels = {}
for parasite in content['parasites']:
    labels[str(parasite)] = '<i>'+content['parasites'][parasite]['label']+'</i>'
```

```python
labels
```

```python
net['taxid1'] = net['taxid1'].astype(str)
```

```python
net = net.replace({"taxid1": labels})
```

```python
net.head()
```

```python
fig = px.box(net.sort_values("taxid1"), x="taxid1", y="average_score", color='taxid1')
fig.update_traces(showlegend=False)
fig
```

```python
net[net['taxid1'] == '<i>Trichinella spiralis</i>'].shape
```

```python
net[net['taxid1'] == '<i>Trichinella spiralis</i>']['source'].unique().shape
```

```python
fig = px.bar(net.groupby(["taxid1"]).count().reset_index().sort_values("taxid2"), x="taxid1", y="average_score", color='taxid1',  labels={"average_score":"count"})
fig
```

```python
aux = net[net["average_score"]>=0.7]
aux = aux.groupby("target_name").count().reset_index().sort_values("taxid1", ascending=False)[0:20]
aux.head()
```

```python
fig = px.bar(aux, x="target_name", y="taxid1", labels={"taxid1":"degree"})
fig.update_traces(showlegend=False)
fig
```

```python
aux = net[net["average_score"]>=0.7]
aux = aux[['taxid1', 'target_name']].drop_duplicates()
aux = aux.groupby("target_name").count().reset_index().sort_values("taxid1", ascending=False)[0:20]
aux.head()
```

```python
fig = px.bar(aux, x="target_name", y="taxid1", labels={"taxid1":"num parasitic species"})
fig.update_traces(showlegend=False)
fig
```

```python
aux = net[net["average_score"]>=0.7]
aux = aux[['taxid1', 'target_name']].drop_duplicates()
aux = aux.groupby("target_name").count().reset_index().sort_values("taxid1", ascending=False)
list_specific = aux[ aux['taxid1']< 2]['target_name'].tolist()

aux = net[net['target_name'].isin(list_specific)]
```

```python
aux.head()
```

```python
aux.groupby("taxid1").count().reset_index().sort_values("taxid1", ascending=False)
```

```python
fig = px.scatter(aux.groupby("taxid1").count().reset_index().sort_values("taxid2", ascending=False), 
                 x="taxid1", y="source", color="taxid1", size="source", labels={"source":"num specific interactions"})
fig.update_traces(showlegend=False)
fig
```
